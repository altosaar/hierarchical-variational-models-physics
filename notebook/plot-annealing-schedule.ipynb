```
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
import numpy as np
```

```
def exponential(step, final_value, rate, max_step):
    return final_value * (1 - rate ** (step / max_step))

def linear(step, final_value, max_step):
  return final_value * min(1, step / max_step)
```

```

steps = np.arange(20000)
fig, ax = plt.subplots(figsize=(3 * 1.618, 3))
ax.plot(steps, [linear(step, 1, 20000) for step in steps], label='linear')
ax.plot(steps, [exponential(step, 1, 0.3, 5000) for step in steps], label='exponential')
ax.legend()
plt.show()
```

```

steps = np.arange(40000)
fig, ax = plt.subplots(figsize=(3 * 1.618, 3))
ax.plot(steps,[20.0 * max(1/20, 0.2 ** (step / 20000)) for step in steps], label='exponential')
ax.legend()
plt.show()
```

```
fig, ax = plt.subplots(figsize=(3 * 1.618, 3))
for rate in [1e-6, 0.01, 0.001, 0.0001]:
  ax.plot(steps, [exponential(step, 1, rate, 5000) for step in steps], label='rate=%.3e' % rate)
ax.plot(steps, [linear(step, 1, 5000) for step in steps], label='linear')
ax.legend()
ax.set(title='exponential schedule')  
```
